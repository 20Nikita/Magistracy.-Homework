# 1) Функции случайных велечин

#### Распределение Рэлeя $\frac{x}{a^2}*e^{\frac{-x^2}{2*a^2}}$ , а = 3

```python
def Rela(x):
    return (x/3**2) * Math.E**(-x**2/(2*3**2))
```

#### Распределение Лаплпсса $y=\frac{a}{2}e^{-a\left|x-b\right|}$ , a = 3, b = 1

```python
def Laplass(x):
    return 3/2 * Math.E**(-3*abs(x-1))
```
