```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n_storis = 1000
max_time = 200
```

```python
t = np.arange(max_time)
steps = 2 * np.random.randint(0,2,(n_storis,max_time)) -1
print(steps)
```

```python
positions = np.cumsum(steps,axis = 1)
```

```python
kv_dist = positions**2
```

```python
min_skvare_distans = np.sqrt(np.mean(kv_dist,axis = 0))
min_skvare_distans
```

```python
plt.plot(t,min_skvare_distans)
```
